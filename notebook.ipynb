### Cleaning Data

Dataset day dan hour memiliki isi yang sama, hanya saja ditungkan dalam basis yang berbeda. Dataset day dituangkan dalam basis per-hari, sedangkan dataset hour dituangkan dalam basis per-jam. Dengan mempertimbangkan pertanyaan bisnis, dataset yang akan digunakan adalah dataset hour, maka dataset day akan dihapus.

```python
# Menghapus dataset day
del day_df
```

instant: tidak memberikan informasi tambahan yang berguna.

weathersit, atemp, windspeed, hum: tidak ada pertanyaan bisnis yang terkait dengan parameter ini.

```python
# Menghapus kolom yang tidak digunakan
drop_col = ['instant', 'weathersit', 'atemp', 'windspeed', 'hum']

for i in hour_df.columns:
  if i in drop_col:
    hour_df.drop(labels=i, axis=1, inplace=True)

hour_df.head()
```

```python
# Mengubah nama kolom pada dataframe agar mudah diidentifikasi
hour_df.rename(columns={'dteday':'datetime',
                        'yr':'year',
                        'mnth' : 'month',
                        'hr' : 'hour',
                        'casual' : 'user_casual',
                        'registered' : 'user_registered',
                        'cnt':'count_total'},
               inplace=True)

hour_df.head()
```

```python
# Mengubah type kolom datetime yang semula bertipe objek menjadi datetime
hour_df['datetime'] = pd.to_datetime(hour_df.datetime)

hour_df.head()
```

```python
# Pada dataset, nilai suhu merupakan hasil standarisasi dengan dibagi dengan nilai maksimal nya yaitu 41 C.
# Untuk mempermudah pemahaman, suhu diubah menjadi nilai aslinya. Dan dibuat kolom baru bernama temperature
hour_df['temperature'] = hour_df['temp'] * 41

# Kemudian untuk menghindari kesalahan, kolom temp akan dihapus
del hour_df['temp']

hour_df
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Mencari info statistik pada setiap kolom yang ada di dataframe
hour_df.describe()
```

```python
# Ekplorasi jumlah user registered dan user casual pada setiap bulan di tahun 2011-2012
hour_df.groupby(['year', 'month'])[['user_registered', 'user_casual']].sum().reset_index()
```

```python
# Membuat variabel data_2011 untuk ekplorasi jumlah user registered dan user casual pada setiap bulan di tahun 2011
data_2011 = hour_df[hour_df['year'] == 0]
monthly_data2011 = data_2011.groupby(by='month').agg({
    'user_casual':'sum',
    'user_registered':'sum'
})
monthly_data2011
```

Penyewaan sepeda tertinggi pada tahun 2011 oleh user casual terjadi pada bulan ke-7 (Juli) sebanyak 36452 user, sedangkan oleh user registered terjadi pada bulan ke-6 (Juni) sebanyak 112900 user.

```python
# Membuat variabel data_2012 untuk ekplorasi jumlah user registered dan user casual pada setiap bulan di tahun 2012
data_2012 = hour_df[hour_df['year'] == 1]
monthly_data2012 = data_2012.groupby(by='month').agg({
    'user_casual':'sum',
    'user_registered':'sum'
})
monthly_data2012
```

Pada tahun 2012, penyewaan sepeda oleh user casual mencapai puncaknya di bulan ke-5 (Mei) sebanyak 44235 user, sedangkan oleh user registered terjadi di bulan ke-9 (September) sebanyak 174795 user.

```python
# Eksplorasi jumlah penyewaan sepeda pada hari libur (holiday)
hour_df.groupby(by='holiday').agg({
    'user_casual':'sum',
    'user_registered':'sum'
})
```

Baik user casual maupun user registered lebih banyak menyewa sepeda pada hari biasa(weekday) atau hari kerja (workingday) daripada di hari libur (holiday). Penurunan jumlah user casual sebanyak 89%, sedangkan pada user registered sebanyak 96%.

```python
# Eksplorasi jumlah penyewaan sepeda pada hari kerja (working day)
hour_df.groupby(by='workingday').agg({
    'user_casual':'sum',
    'user_registered':'sum'
})
```

Meskipun perbedaan jumlahnya tidak terlalu signifikan, user casual lebih banyak menyewa sepeda di hari libur daripada di hari kerja (working day). Namun sebaliknya, user registered lebih banyak menyewa sepeda pada hari kerja (working day)

```python
# Eksplorasi jumlah penyewaan sepeda berdasarkan hari
hour_df.groupby(by='weekday').agg({
    'user_casual':'sum',
    'user_registered':'sum'
})
```

Hari Sabtu merupakan hari favorit user casual untuk menyewa sepeda, sedangkan user registered hari favoritnya adalah hari Kamis. Hal itu ditunjukkan dengan jumlah penyewa sepeda pada hari yang dimaksud.

```python
# Eksplorasi jumlah penyewaan sepeda pada setiap jam(hour)
hour_df.groupby(by=['hour']).agg({
    'user_casual':'sum',
    'user_registered':'sum'

})
```

Penyewaan sepeda oleh user casual paling sedikit pada jam 4 pagi, dan paling banyak pada jam 2 sore. Sedangkan penyewaan sepeda oleh user registered paling sedikit pada jam 4 pagi, dan paling banyak pada jam 5 sore.

```python
# Eksplorasi jumlah penyewaan sepeda pada setiap musim(season)
hour_df.groupby(by=['season']).agg({
    'user_casual':'sum',
    'user_registered':'sum'

})
```

Baik user casual maupun user registered cenderung sedikit menyewa sepeda pada musim ke-1 yaitu musim semi (springer), dan cenderung banyak menyewa sepeda pada musim gugur ke-3 yaitu musim (fall).

```python
# Membuat bin(interval) untuk kolom temperatur yang nilainya berupa nilai minimal, nilai quartil (25, 50, 75), serta nilai maksimal nya
bins_temp = [0.82, 13.94, 20.50, 27.06, 41.00]

# Mengelompokkan data ke dalam bin
hour_df['temp_group'] = pd.cut(hour_df['temperature'], bins=bins_temp, labels=['0.82-13.94','13.94-20.50', '20.50-27.06', '27.06-41.00'])

# Eksplorasi jumlah penyewaan sepeda pada setiap kelompok suhu (temp_group)
hour_df.groupby(by=['temp_group']).agg({
    'user_casual':'sum',
    'user_registered':'sum'

})
```

Pada range suhu yang rendah antara 0.82-13.94 C user casual dan user registered cenderung sedikit menyewa sepeda. Sebaliknya, pada range suhu yang tinggi antara 27.06-41 C user casual dan registered cenderung lebih banyak menyewa sepeda.

### PREDIKSI MENGGUNAKAN MOVING AVERAGE
Moving Average (MA), atau yang dikenal juga sebagai Rata-rata Bergerak, adalah metode statistik yang digunakan untuk merata-ratakan fluktuasi data sepanjang waktu dengan menggunakan jendela (window) tertentu. Dalam hal ini menggunakan 3 jendela, yaitu setiap nilai dalam kolom akan dihitung sebagai rata-rata dari nilai-nilai sekitarnya, yaitu nilai itu sendiri dan dua nilai sebelumnya. Tujuan utamanya adalah untuk mengidentifikasi tren atau pola dalam data.

```python
# Menambahkan kolom baru untuk moving average pada total penggunaan (count_total)
hour_df['moving_avg'] = hour_df['count_total'].rolling(window=3).mean()

# Menggunakan matplotlib untuk membuat plot dari data asli dan moving average
plt.figure(figsize=(12, 6))
plt.plot(hour_df['datetime'], hour_df['count_total'], label='Actual', marker='o')
plt.plot(hour_df['datetime'], hour_df['moving_avg'], label='Moving Average', marker='o')

# Menambahkan label dan judul
plt.xlabel('Tanggal')
plt.ylabel('Total Penggunaan Sepeda')
plt.title('Prediksi Menggunakan Moving Average')

# Menambahkan legenda untuk membedakan antara data asli dan moving average
plt.legend()

# Menampilkan plot
plt.show()
```

Hasil analisis menggunakan moving average dari data penyewaan sepeda tahun 2011-2012, menghasilkan prediksi penyewaan sepeda pada tahun berikutnya yaitu tahun 2013. Dapat dilihat nilai moving average dengan nilai asli (actual) memiliki perbedaan yang tidak terlalu signifikan. Hal ini dapat memperkuat prediksi yang dihasilkan.
